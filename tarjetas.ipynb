## MACHINE LEARNING
###### Con el uso de Machine Learning identificaremos si efectivamente los resultados obtenidos en el analisis realizado anteriormente coinciden con los resultados que obtendremos con las tecnicas que utilizaremos en este módulo.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('tarjetas.csv')
df
```

```python
x = df[['monto_deuda', 'meses_inactivos_12meses', 'limite_credito', 'num_cliente']]
```

```python
# Gráfica de sedimentación para determinar el número de clústeres
ks = range(1, 11)
inertias = []

for k in ks:
    model = KMeans(n_clusters = k)
    model.fit(x)
    inertias.append(model.inertia_)
    
# Gráfico de sedimentación para ver las incercias
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

###### Se observa que en la grafica de sedimentacion hasta la iteracion numero 3 ya no hay cambios en la inercia de manera significativa por lo que el numero de clusters debe ser de 3 ya que a partir del punto 3 la inercia ya no esta cambiando.
###### A pesar del resultado que nos arroja la grafica de sedimentacion de tomar 3 clusters, unicamente seleccionare 2 ya que estamos pronosticando 2 opciones: cuales serian los clientes que abandoran el uso de la tarjeta de credito y cuales clientes no abandonaran el uso de las tarjetas de credito.

```python
# Definimos el número de centroides determinado
K = 2

# Determinamos el número de centróides y asignamos los datos
modelo_kmeans = KMeans(n_clusters = K)

modelo_kmeans.fit(x)
labels = modelo_kmeans.fit_predict(x)

#Y extraemos los centroides.
centroides = modelo_kmeans.cluster_centers_
print(centroides)
```
